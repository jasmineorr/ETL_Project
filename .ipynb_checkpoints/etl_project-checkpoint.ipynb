```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
corona_file = "Resources/2019_nCoV_data.csv"
corona_data_df = pd.read_csv(corona_file)
corona_data_df.head()
```

```python
new_corona_data_df = corona_data_df[['Country', 'Confirmed', 'Deaths']].copy()
new_corona_data_df.head()
```

```python
grouped_corona_data_df = new_corona_data_df.groupby(['Country'], as_index=False).sum()
grouped_corona_data_df
```

```python
updated_corona_data_df = grouped_corona_data_df.rename(columns = {'Confirmed':'Cases'})
updated_corona_data_df
```

```python
replaced_corona_data_df = updated_corona_data_df.replace({'Mainland China': 'China'})
replaced_corona_data_df
```

```python
sorted_corona_df = replaced_corona_data_df.sort_values(by=['Country'], ascending=True)
sorted_corona_df
```

```python
reindexed_corona_df = sorted_corona_df.reset_index(drop=True)
reindexed_corona_df
```
